```python
#Problem 1d
import numpy as np
import matplotlib.pyplot as plt
```

```python
#define x range
x = np.linspace(0, 10, 1000) 

#optimal a*
a_star = np.sqrt(3) - 1

#define densities
f = (1/3) * x * (1 + x) * np.exp(-x)
g = a_star**2 * x * np.exp(-a_star * x)

#envelope constant
c = np.exp(-a_star) / (3 * a_star**2 * (1 - a_star))

# Plot
plt.figure()
plt.plot(x, f, label="f(x)")
plt.plot(x, c * g, label=r"$c(a^*) g_{a^*}(x)$")

plt.xlabel("x")
plt.ylabel("Density")
plt.legend()
plt.title("Target density and scaled proposal")
plt.savefig("pbm1d.png", dpi=300, bbox_inches="tight")

plt.show()
```
